# Задача 1

Напишите программу, которая считывает одну строку. Если это строка «Искусственный Интеллект», программа выводит «ДА», в противном случае программа выводит «НЕТ»

Для этой задачи вам нужно использовать функции Python: `input()`, `print()`, `if` и `else`.

Пример вывода программы:


```

```

```
Блокчейн
НЕТ
```


**Формат выходных данных**

# Задача 8

Даны три целых числа. Определите, сколько среди них совпадающих. Программа должна вывести одно из чисел: 3 (если все совпадают), 2 (если два совпадает) или 0 (если все числа различны).


Решите данную задачу тремя способами:
* 1 способ. Использование вложенного условного оператора.


  ```
  if условие1:
      блок кода
  else:
      if условие2:
          блок кода
      else:
          if условие3:
              блок кода  
  ```
* 2 способ. Использование каскадного условного оператора.


```
  if условие1:
      блок кода
  elif условие2:
      блок кода
  elif условие3:
      блок кода
  ...
  else:
      блок кода
```

* 3 способ. Использование каскадного условного оператора и логического оператора `or`



Пример вывода программы:


```
Число 1: 1
Число 2: 1
Число 3: 1
Совпадают: 3
```




```
Число 1: 1
Число 2: 2
Число 3: 1
Совпадают: 2
```

```
Число 1: 1
Число 2: 2
Число 3: 3
Совпадают: 0
```

````python
#1 способ. Использование вложенного условного оператора

i = int(input("число 1: "))
k = int(input("число 2: "))
j = int(input("число 3: "))

if i == k == j :
  print("Совпадают: 3")
else:
    if i == k != j :
      print("Совпадают: 2")
    else:
      if i == j != k or k == j != i:
       print("Совпадают: 2")
      else:
         if i != k != j != i :
           print("Совпадают: 0")
````

````python
#2 способ. Использование каскадного условного оператора.

i = int(input("число 1: "))
k = int(input("число 2: "))
j = int(input("число 3: "))

if i == k == j :
  print("Совпадают: 3")
elif i == k != j :
      print("Совпадают: 2")
elif i == j != k or k == j != i:
      print("Совпадают: 2")
else:
   if i != k != j :
     print("Совпадают: 0")
````

````python
#3 способ. Использование каскадного условного оператора и логического оператора or

i = int(input("число 1: "))
k = int(input("число 2: "))
j = int(input("число 3: "))

if i == k == j :
  print("Совпадают: 3")
else:
    if i == k != j or i == j != k or k == j != i :
      print("Совпадают: 2")
    else:
      if i != k != j != i :
         print("Совпадают: 0")

````

* Если ни одно из условий не выполняется, выведите соответствующее сообщение: `"Добавьте больше товаров в корзину и поменяйте товар на фрукт, чтобы получить скидку 20%."`

Для этой задачи вам нужно использовать функции Python: `input()`, `print()`, `if`,`elif`, `else`, `int()`, `>`, `==`, `and`.
